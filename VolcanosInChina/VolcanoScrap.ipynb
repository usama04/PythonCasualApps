```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://ngdc.noaa.gov/nndc/struts/results?type_0=Like&query_0=&op_8=eq&v_8=China&type_10=EXACT&query_10=None+Selected&le_2=&ge_3=&le_3=&ge_2=&op_5=eq&v_5=&op_6=eq&v_6=&op_7=eq&v_7=&t=102557&s=5&d=5"
headers = {"user-agent":"Mozilla/5.0"}
```

```python
r = requests.get(url, headers=headers)
```

```python
r.status_code
```

```python
c = r.content
#with open('/home/usama/Desktop/volcanoes.html', 'r') as volfile:
#    c = volfile.read()
soup = BeautifulSoup(c, "html.parser")
soup.prettify()
```

```python
tables = soup.find_all("table", {'summary':'layout table'})
len(tables)
```

```python
for table in tables:
    print(table)
    print("\n")
```

```python
table = tables[1]
#table=tables[0]
print(table)
```

```python
columns = []
heads = table.find('thead')
values = heads.find_all('th')
for value in values:
    print(value.text + "\n")
    columns.append(value.text)
```

```python
rows = table.find_all('tr')
len(rows)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns=columns)
```
